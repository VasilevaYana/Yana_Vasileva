```python
import pandas as pd
import numpy as np
import random
import os

# === ВАЖНО: Фиксация Random Seed (по правилам соревнования) ===
def seed_everything(seed=993):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    # Если позже подключим torch, добавим сюда фиксацию для него

seed_everything()
print("Seed 993 зафиксирован.")

# === Загрузка данных ===
TRAIN_PATH = '/content/train.csv'
TEST_PATH = '/content/test.csv'
SUB_PATH = '/content/submission.csv'

try:
    df_train = pd.read_csv(TRAIN_PATH)
    df_test = pd.read_csv(TEST_PATH)
    df_sub = pd.read_csv(SUB_PATH)

    print("\n--- Размеры данных ---")
    print(f"Train shape: {df_train.shape}")
    print(f"Test shape:  {df_test.shape}")

    print("\n--- Пример данных (Train) ---")
    display(df_train.head(2))

    print("\n--- Распределение целевой переменной (Relevance) ---")
    print(df_train['relevance'].value_counts().sort_index())

    print("\n--- Какие языки/локали есть (Top 5) ---")
    print(df_train['product_locale'].value_counts().head(5))

    print("\n--- Проверка на пропуски (Train) ---")
    print(df_train.isnull().sum())

except FileNotFoundError:
    print("Ошибка! Файлы не найдены. Проверь, что загрузила их в папку /content/ слева.")
```

```python
# === 1. Заполнение пропусков ===
# Список текстовых колонок
text_cols = ['query', 'product_title', 'product_description', 'product_bullet_point', 'product_brand', 'product_color']

for col in text_cols:
    df_train[col] = df_train[col].fillna("").astype(str)
    df_test[col] = df_test[col].fillna("").astype(str)

print("Пропуски заполнены.")

# Проверим локали в тесте, вдруг там сюрприз
print("Локали в тесте:\n", df_test['product_locale'].value_counts())

# === 2. Генерация простых признаков (Feature Engineering) ===

def create_features(df):
    # Длины строк
    df['query_len'] = df['query'].apply(len)
    df['title_len'] = df['product_title'].apply(len)
    df['desc_len'] = df['product_description'].apply(len)

    # Функция для подсчета пересечения слов
    def word_overlap(row):
        # Разбиваем запрос и заголовок на слова (set убирает дубликаты)
        q_words = set(row['query'].lower().split())
        t_words = set(row['product_title'].lower().split())
        # Считаем общие слова
        common = q_words.intersection(t_words)
        return len(common)

    df['word_overlap'] = df.apply(word_overlap, axis=1)

    return df

print("Генерация признаков для Train...")
df_train = create_features(df_train)

print("Генерация признаков для Test...")
df_test = create_features(df_test)

# Посмотрим на новые признаки
print("\n--- Пример новых признаков ---")
print(df_train[['query', 'product_title', 'query_len', 'word_overlap']].head(3))
```

```python
# Установка CatBoost (если нет в Colab)
!pip install catboost -q

from catboost import CatBoostRanker, Pool
from sklearn.model_selection import GroupShuffleSplit

# 1. Подготовка данных
# Выберем признаки для обучения (пока только числовые)
feature_cols = ['query_len', 'title_len', 'desc_len', 'word_overlap']
target_col = 'relevance'
group_col = 'query_id'

# Сортируем данные по query_id (требование CatBoost Ranker)
df_train = df_train.sort_values(by='query_id').reset_index(drop=True)
df_test = df_test.sort_values(by='query_id').reset_index(drop=True)

# 2. Разбиение на Train и Validation по Группам (Query ID)
# Мы берем 80% запросов в обучение, 20% в валидацию
gss = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=993)
train_idx, val_idx = next(gss.split(df_train, groups=df_train['query_id']))

train_data = df_train.iloc[train_idx]
val_data = df_train.iloc[val_idx]

print(f"Размер Train: {train_data.shape}, Размер Val: {val_data.shape}")

# 3. Создание Pool для CatBoost
# Pool - это специальный формат данных для этой библиотеки
train_pool = Pool(
    data=train_data[feature_cols],
    label=train_data[target_col],
    group_id=train_data[group_col] # Группировка по запросам
)

val_pool = Pool(
    data=val_data[feature_cols],
    label=val_data[target_col],
    group_id=val_data[group_col]
)

test_pool = Pool(
    data=df_test[feature_cols],
    group_id=df_test[group_col]
)

# 4. Настройка и обучение модели
# Используем YetiRank - один из лучших лоссов для ранжирования
model = CatBoostRanker(
    iterations=1000,           # Количество деревьев
    learning_rate=0.05,        # Скорость обучения
    depth=6,                   # Глубина деревьев
    loss_function='YetiRank',  # Оптимизация ранжирования
    eval_metric='NDCG:top=10', # Метрика соревнования
    random_seed=993,           # Фиксация сида
    verbose=100                # Вывод логов каждые 100 итераций
)

print("\nНачинаем обучение CatBoost...")
model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=100)

# 1. Правильный вывод важности признаков (передаем train_pool)
print("\nВажность признаков (LossFunctionChange):")
try:
    # Для ранжирования нужно передавать пул данных
    print(model.get_feature_importance(data=train_pool, prettified=True))
except Exception as e:
    print(f"Не удалось посчитать важность (не критично): {e}")

# 2. Предсказание на Test
print("\nДелаем предсказание...")
preds = model.predict(test_pool)

# 3. Сохранение submission
submission = df_test[['id']].copy()
submission['prediction'] = preds

# Сохраняем в CSV
submission.to_csv('submission.csv', index=False)
print("\nФайл submission.csv успешно сохранен! Можно скачивать.")

# Показать первые строки для проверки
print(submission.head())
```

```python
# Установка библиотеки
!pip install sentence_transformers -q

from sentence_transformers import SentenceTransformer
import torch
import numpy as np

# Проверка GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Используем устройство: {device}")

# Загружаем модель (она маленькая и быстрая)
st_model = SentenceTransformer('all-MiniLM-L6-v2', device=device)

def get_embeddings_similarity(df, model, batch_size=64):
    # 1. Получаем эмбеддинги запросов
    print("Кодируем запросы...")
    query_embeddings = model.encode(
        df['query'].values,
        batch_size=batch_size,
        convert_to_tensor=True,
        show_progress_bar=True
    )

    # 2. Получаем эмбеддинги заголовков
    print("Кодируем заголовки...")
    title_embeddings = model.encode(
        df['product_title'].values,
        batch_size=batch_size,
        convert_to_tensor=True,
        show_progress_bar=True
    )

    # 3. Считаем косинусное сходство (Cosine Similarity)
    print("Считаем сходство...")
    # Cosine Similarity = (A . B) / (|A| * |B|)
    # sentence_transformers выдает уже нормализованные векторы, так что просто точечное произведение
    # Используем torch для скорости на GPU
    cos_scores = torch.nn.functional.cosine_similarity(query_embeddings, title_embeddings)

    return cos_scores.cpu().numpy()

# === Применяем к Train и Test ===
print("--- Обработка Train ---")
df_train['bert_cosine'] = get_embeddings_similarity(df_train, st_model)

print("\n--- Обработка Test ---")
df_test['bert_cosine'] = get_embeddings_similarity(df_test, st_model)

# Проверим, как это выглядит
print("\nПример:")
print(df_train[['query', 'product_title', 'word_overlap', 'bert_cosine']].head(5))
```

```python
# === Полный блок: BM25 + Бренды + Обучение ===
!pip install rank_bm25 -q

from rank_bm25 import BM25Okapi
from catboost import CatBoostRanker, Pool
from sklearn.model_selection import GroupShuffleSplit
import numpy as np
import pandas as pd

# 1. Генерация BM25 фичей
# Объединяем тексты для обучения словаря
all_titles = pd.concat([df_train['product_title'], df_test['product_title']]).tolist()
tokenized_corpus = [doc.split(" ") for doc in all_titles]

print("Строим индекс BM25...")
bm25 = BM25Okapi(tokenized_corpus)
word_idf = bm25.idf

def fast_bm25_score(row):
    # Разбиваем на слова
    q_words = set(row['query'].lower().split())
    t_words = set(row['product_title'].lower().split())
    # Ищем общие слова
    common = q_words.intersection(t_words)
    score = 0.0
    for w in common:
        try:
            # Суммируем вес (редкость) каждого общего слова
            score += word_idf.get(w, 0)
        except:
            pass
    return score

print("Считаем BM25 для Train...")
df_train['bm25_score'] = df_train.apply(fast_bm25_score, axis=1)

print("Считаем BM25 для Test...")
df_test['bm25_score'] = df_test.apply(fast_bm25_score, axis=1)

# 2. Обработка категориальных признаков (Бренд)
df_train['product_brand'] = df_train['product_brand'].fillna("unknown").astype(str)
df_test['product_brand'] = df_test['product_brand'].fillna("unknown").astype(str)

# 3. Подготовка данных для CatBoost
# Список всех фичей (должен включать bert_cosine с прошлого шага)
feature_cols = ['query_len', 'title_len', 'desc_len', 'word_overlap', 'bert_cosine', 'bm25_score', 'product_brand']
cat_features = ['product_brand']
target_col = 'relevance'
group_col = 'query_id'

# Сортировка по группам (обязательно для Learning to Rank)
df_train = df_train.sort_values(by='query_id').reset_index(drop=True)
df_test = df_test.sort_values(by='query_id').reset_index(drop=True)

# Разбиение на Train/Val (строго по группам)
gss = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=993)
train_idx, val_idx = next(gss.split(df_train, groups=df_train['query_id']))

# === СОЗДАНИЕ ПУЛОВ (ИСПРАВЛЕНО) ===
train_pool = Pool(
    data=df_train.iloc[train_idx][feature_cols],
    label=df_train.iloc[train_idx][target_col],
    group_id=df_train.iloc[train_idx][group_col],
    cat_features=cat_features
)

val_pool = Pool(
    data=df_train.iloc[val_idx][feature_cols],
    label=df_train.iloc[val_idx][target_col],
    group_id=df_train.iloc[val_idx][group_col], # <-- ТЕПЕРЬ ТУТ ПРАВИЛЬНЫЙ ИНДЕКС val_idx
    cat_features=cat_features
)

test_pool = Pool(
    data=df_test[feature_cols],
    group_id=df_test[group_col],
    cat_features=cat_features
)

# 4. Обучение модели
model = CatBoostRanker(
    iterations=2500,
    learning_rate=0.05,
    depth=6,
    loss_function='YetiRank',
    eval_metric='NDCG:top=10',
    random_seed=993,
    verbose=100,
    task_type="GPU" # Можно раскомментировать, если есть GPU, но для таких данных CPU справится быстро
)

print("\nНачинаем обучение...")
model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=100)

# 5. Анализ важности
print("\nВажность признаков:")
print(model.get_feature_importance(data=train_pool, prettified=True))

# 6. Сабмит
submission = df_test[['id']].copy()
submission['prediction'] = model.predict(test_pool)
submission.to_csv('submission_bm25_brand.csv', index=False)
print("\nФайл submission_bm25_brand.csv успешно сохранен!")
```

```python
from sklearn.model_selection import GroupKFold
import gc # Сборщик мусора, чтобы не забить память

# Настройки
N_FOLDS = 5
feature_cols = ['query_len', 'title_len', 'desc_len', 'word_overlap', 'bert_cosine', 'bm25_score', 'product_brand']
cat_features = ['product_brand']
target_col = 'relevance'
group_col = 'query_id'

# Инициализируем массив для хранения предсказаний на тесте (заполнен нулями)
final_predictions = np.zeros(len(df_test))

# Стратегия разбиения (5 фолдов)
gkf = GroupKFold(n_splits=N_FOLDS)

print(f"Запускаем обучение на {N_FOLDS} фолдах...")

# Цикл по фолдам
for fold, (train_idx, val_idx) in enumerate(gkf.split(df_train, groups=df_train[group_col])):
    print(f"\n=== FOLD {fold+1}/{N_FOLDS} ===")

    # Формируем пулы для текущего фолда
    train_pool = Pool(
        data=df_train.iloc[train_idx][feature_cols],
        label=df_train.iloc[train_idx][target_col],
        group_id=df_train.iloc[train_idx][group_col],
        cat_features=cat_features
    )

    val_pool = Pool(
        data=df_train.iloc[val_idx][feature_cols],
        label=df_train.iloc[val_idx][target_col],
        group_id=df_train.iloc[val_idx][group_col],
        cat_features=cat_features
    )

    # Создаем новую модель
    model = CatBoostRanker(
        iterations=2000,
        learning_rate=0.05,
        depth=6,
        loss_function='YetiRank',
        eval_metric='NDCG:top=10',
        random_seed=993 + fold, # Меняем сид для каждого фолда для разнообразия
        verbose=200
    )

    # Обучаем
    model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=100)

    # Предсказываем на тесте
    # Важно: суммируем предсказания
    final_predictions += model.predict(test_pool)

    # Чистим память
    del model, train_pool, val_pool
    gc.collect()

# Усредняем предсказания (делим сумму на 5)
final_predictions /= N_FOLDS

# Сохраняем результат
submission = df_test[['id']].copy()
submission['prediction'] = final_predictions
submission.to_csv('submission_cv5.csv', index=False)

print("\nОбучение завершено!")
print("Файл submission_cv5.csv готов к отправке.")
```

```python
# Чистим память от старых моделей
import gc
import torch
try:
    del st_model
    del model
    gc.collect()
    torch.cuda.empty_cache()
except:
    pass

from sentence_transformers import SentenceTransformer

# 1. Загружаем более мощную модель (MPNet)
print("Загружаем модель all-mpnet-base-v2 (она мощнее)...")
st_model = SentenceTransformer('all-mpnet-base-v2', device='cuda')

# 2. Создаем расширенный текст для товаров
# Мы берем Заголовок + первые 200 символов описания + буллеты
# Это даст модели больше контекста
def make_full_text(row):
    # Ограничиваем описание, чтобы не было слишком длинно
    desc = row['product_description'][:200]
    bullets = row['product_bullet_point'][:200]
    return f"{row['product_title']} {desc} {bullets}"

print("Подготовка текстов (Title + Desc)...")
# Применяем и сразу превращаем в список для ускорения
train_texts = df_train.apply(make_full_text, axis=1).tolist()
test_texts = df_test.apply(make_full_text, axis=1).tolist()
train_queries = df_train['query'].tolist()
test_queries = df_test['query'].tolist()

# 3. Пересчитываем косинусное сходство с новой моделью и данными
def get_embeddings_similarity_optimized(queries, docs, model, batch_size=64):
    # Кодируем запросы
    print("Кодируем запросы...")
    q_emb = model.encode(queries, batch_size=batch_size, convert_to_tensor=True, show_progress_bar=True)

    # Кодируем документы
    print("Кодируем товары...")
    d_emb = model.encode(docs, batch_size=batch_size, convert_to_tensor=True, show_progress_bar=True)

    # Считаем сходство
    print("Считаем cosine score...")
    return torch.nn.functional.cosine_similarity(q_emb, d_emb).cpu().numpy()

print("\n--- Генерируем новые эмбеддинги для Train ---")
df_train['bert_cosine'] = get_embeddings_similarity_optimized(train_queries, train_texts, st_model)

print("\n--- Генерируем новые эмбеддинги для Test ---")
df_test['bert_cosine'] = get_embeddings_similarity_optimized(test_queries, test_texts, st_model)

# 4. Обучаем CatBoost (снова 5 фолдов, чтобы было надежно)
from catboost import CatBoostRanker, Pool
from sklearn.model_selection import GroupKFold
import numpy as np

# Настройки те же
N_FOLDS = 5
feature_cols = ['query_len', 'title_len', 'desc_len', 'word_overlap', 'bert_cosine', 'bm25_score', 'product_brand']
cat_features = ['product_brand']
target_col = 'relevance'
group_col = 'query_id'

final_predictions = np.zeros(len(df_test))
gkf = GroupKFold(n_splits=N_FOLDS)

print(f"\nЗапускаем обучение CV-{N_FOLDS} на новых фичах...")

for fold, (train_idx, val_idx) in enumerate(gkf.split(df_train, groups=df_train[group_col])):
    print(f"Fold {fold+1}...")

    train_pool = Pool(df_train.iloc[train_idx][feature_cols], label=df_train.iloc[train_idx][target_col],
                      group_id=df_train.iloc[train_idx][group_col], cat_features=cat_features)
    val_pool = Pool(df_train.iloc[val_idx][feature_cols], label=df_train.iloc[val_idx][target_col],
                    group_id=df_train.iloc[val_idx][group_col], cat_features=cat_features)

    model = CatBoostRanker(iterations=2000, learning_rate=0.05, depth=6, loss_function='YetiRank',
                           eval_metric='NDCG:top=10', random_seed=993 + fold, verbose=0) # verbose=0 чтоб не засорять экран

    model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=50)
    print(f"  Best Validation NDCG: {model.get_best_score()['validation']['NDCG:top=10']:.5f}")

    final_predictions += model.predict(Pool(df_test[feature_cols], group_id=df_test[group_col], cat_features=cat_features))

    # Чистим память внутри цикла
    del model, train_pool, val_pool
    gc.collect()

final_predictions /= N_FOLDS

# Сабмит
submission = df_test[['id']].copy()
submission['prediction'] = final_predictions
submission.to_csv('submission_mpnet_desc.csv', index=False)
print("\nФайл submission_mpnet_desc.csv готов! Это наша самая мощная попытка.")
```

```python
from catboost import CatBoostRanker, Pool
from sklearn.model_selection import GroupKFold
import numpy as np
import gc

# Настройки
N_FOLDS = 5
feature_cols = ['query_len', 'title_len', 'desc_len', 'word_overlap', 'bert_cosine', 'bm25_score', 'product_brand']
cat_features = ['product_brand']
target_col = 'relevance'
group_col = 'query_id'

# Инициализируем массив предсказаний
final_predictions = np.zeros(len(df_test))
gkf = GroupKFold(n_splits=N_FOLDS)

print(f"Перезапускаем обучение CV-{N_FOLDS} (на готовых эмбеддингах)...")

for fold, (train_idx, val_idx) in enumerate(gkf.split(df_train, groups=df_train[group_col])):
    print(f"\nFold {fold+1}...")

    # Создаем пулы
    train_pool = Pool(df_train.iloc[train_idx][feature_cols],
                      label=df_train.iloc[train_idx][target_col],
                      group_id=df_train.iloc[train_idx][group_col],
                      cat_features=cat_features)

    val_pool = Pool(df_train.iloc[val_idx][feature_cols],
                    label=df_train.iloc[val_idx][target_col],
                    group_id=df_train.iloc[val_idx][group_col],
                    cat_features=cat_features)

    # Обучаем
    model = CatBoostRanker(iterations=2000, learning_rate=0.05, depth=6, loss_function='YetiRank',
                           eval_metric='NDCG:top=10', random_seed=993 + fold, verbose=0)

    model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=50)

    # БЕЗОПАСНЫЙ ВЫВОД СКОРА
    # Вместо обращения по ключу, просто выведем весь словарь или пропустим, если ключа нет
    best_scores = model.get_best_score()
    # Пытаемся найти ключ валидации (иногда это 'validation', иногда 'validation_0')
    val_score_key = next((k for k in best_scores.keys() if 'valid' in k), None)

    if val_score_key:
        score = best_scores[val_score_key].get('NDCG:top=10', 0)
        print(f"  Best NDCG: {score:.5f}")
    else:
        print(f"  Модель обучена. (Keys: {best_scores.keys()})")

    # Предсказание
    final_predictions += model.predict(Pool(df_test[feature_cols], group_id=df_test[group_col], cat_features=cat_features))

    # Чистим память
    del model, train_pool, val_pool
    gc.collect()

# Усреднение
final_predictions /= N_FOLDS

# Сохранение
submission = df_test[['id']].copy()
submission['prediction'] = final_predictions
submission.to_csv('submission_mpnet_desc.csv', index=False)
print("\nФайл submission_mpnet_desc.csv готов!")
```

```python
from sentence_transformers import CrossEncoder
import torch
import gc
import numpy as np

# Чистим память от предыдущей тяжелой модели
try:
    del st_model
    torch.cuda.empty_cache()
    gc.collect()
except:
    pass

# 1. Загружаем Cross-Encoder
# Эта модель специально училась отвечать на вопрос: "Подходит ли этот текст этому запросу?"
print("Загружаем Cross-Encoder (ms-marco-MiniLM-L-6-v2)...")
cross_model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2', device='cuda')

# 2. Подготовка пар для оценки
# Мы будем оценивать (Query, Title + Description)
# Ограничим длину описания, чтобы влезло в модель
def prepare_pairs(df):
    pairs = []
    for idx, row in df.iterrows():
        # Текст товара: Заголовок + немного описания
        # Cross-Encoder имеет лимит токенов (обычно 512), лишнее обрежет сам
        item_text = f"{row['product_title']} {str(row['product_description'])[:100]}"
        pairs.append([row['query'], item_text])
    return pairs

print("Готовим пары для Train...")
train_pairs = prepare_pairs(df_train)

print("Готовим пары для Test...")
test_pairs = prepare_pairs(df_test)

# 3. Предсказание скоров (батчами)
batch_size = 128 # Можно больше, модель легкая

print("Считаем Cross-Encoder скоры для Train...")
df_train['cross_score'] = cross_model.predict(train_pairs, batch_size=batch_size, show_progress_bar=True)

print("Считаем Cross-Encoder скоры для Test...")
df_test['cross_score'] = cross_model.predict(test_pairs, batch_size=batch_size, show_progress_bar=True)

# Проверим, как выглядит новый признак
print("\nПример:")
print(df_train[['query', 'product_title', 'cross_score', 'relevance']].head(5))

# === ОБУЧЕНИЕ С НОВЫМ ПРИЗНАКОМ ===
from catboost import CatBoostRanker, Pool
from sklearn.model_selection import GroupKFold

# Добавляем cross_score в список фичей
# У нас теперь очень мощный набор: BERT cosine, MPNet cosine, BM25, Cross-Encoder score
feature_cols = ['query_len', 'title_len', 'desc_len', 'word_overlap', 'bert_cosine', 'bm25_score', 'product_brand', 'cross_score']
cat_features = ['product_brand']
target_col = 'relevance'
group_col = 'query_id'

N_FOLDS = 5
final_predictions = np.zeros(len(df_test))
gkf = GroupKFold(n_splits=N_FOLDS)

print(f"\nЗапускаем финальное обучение с Cross-Encoder фичей...")

for fold, (train_idx, val_idx) in enumerate(gkf.split(df_train, groups=df_train[group_col])):
    print(f"Fold {fold+1}...")

    train_pool = Pool(df_train.iloc[train_idx][feature_cols],
                      label=df_train.iloc[train_idx][target_col],
                      group_id=df_train.iloc[train_idx][group_col],
                      cat_features=cat_features)

    val_pool = Pool(df_train.iloc[val_idx][feature_cols],
                    label=df_train.iloc[val_idx][target_col],
                    group_id=df_train.iloc[val_idx][group_col],
                    cat_features=cat_features)

    # Чуть увеличим learning_rate, так как фича очень сильная
    model = CatBoostRanker(iterations=2000, learning_rate=0.08, depth=6, loss_function='YetiRank',
                           eval_metric='NDCG:top=10', random_seed=993 + fold, verbose=0)

    model.fit(train_pool, eval_set=val_pool, early_stopping_rounds=50)

    final_predictions += model.predict(Pool(df_test[feature_cols], group_id=df_test[group_col], cat_features=cat_features))

    # Важность признаков для первого фолда (любопытно глянуть)
    if fold == 0:
        print("Важность признаков:")
        print(model.get_feature_importance(data=train_pool, prettified=True))

final_predictions /= N_FOLDS

# Сабмит
submission = df_test[['id']].copy()
submission['prediction'] = final_predictions
submission.to_csv('submission_cross_encoder.csv', index=False)
print("\nФайл submission_cross_encoder.csv готов! Надеемся на 0.82+ 🚀")
```
